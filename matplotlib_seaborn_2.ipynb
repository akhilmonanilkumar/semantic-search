```python
z = [1,2,3,4,0.5]

veh_type = ['bicycle','motorbike','car','van','stroler']

plt.pie(z, labels=veh_type)

plt.show()
```

```python
address = 'mtcars.csv'

cars = pd.read_csv(address)

cars.columns = ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']
mpg = cars.mpg

fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

mpg.plot()

ax.set_xlabel('car_names')
ax.set_xticks(range(32))

ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium')
ax.set_title('Miles per Gallon in MT Cars')

ax.set_xlabel('car_names')
ax.set_ylabel('miles/gal')



```

```python
plt.pie(z)

plt.legend(veh_type, loc='best')

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

mpg.plot()

ax.set_xlabel('car_names')
ax.set_xticks(range(32))

ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium')
ax.set_title('Miles per Gallon in MT Cars')

ax.set_xlabel('car_names')
ax.set_ylabel('miles/gal')

ax.set_ylim([0,45])

ax.annotate('Toyota Corolla', xy=(19,33.9), xytext=(21,35), arrowprops=dict(facecolor='black', shrink=0.05))

ax.legend(loc='best')
```
