```python
import pandas as pd 
import numpy as np

import matplotlib.pyplot as plt 
import seaborn as sns
from pylab import rcParams

import scipy
from scipy.stats import spearmanr
```

```python
%matplotlib inline
rcParams['figure.figsize']=14,10
sns.set_style('whitegrid')
```

```python
address = 'mtcars.csv'

cars = pd.read_csv(address)

cars.columns = ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']

cars.head()
```

```python
x= cars[['cyl','vs','am', 'gear']]

sns.pairplot(x)
```

```python
cyl = cars['cyl']

vs=cars['vs']
am=cars['am']
gear=cars['gear']

spearmenr_coefficient, p_value = spearmanr(cyl,vs)
print('SpearmenR Correlation Coefficient %0.3f' % (spearmenr_coefficient))
```

```python
spearmenr_coefficient, p_value = spearmanr(cyl,am)
print('SpearmenR Correlation Coefficient %0.3f' % (spearmenr_coefficient))
```

```python
spearmenr_coefficient, p_value = spearmanr(cyl,gear)
print('SpearmenR Correlation Coefficient %0.3f' % (spearmenr_coefficient))
```

```python
table = pd.crosstab(cyl,am)
from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(table.values)
print('Chi-square Statistics %0.3f p_value %0.3f' % (chi2, p))
```

```python
table = pd.crosstab(cyl,vs)

chi2, p, dof, expected = chi2_contingency(table.values)
print('Chi-square Statistics %0.3f p_value %0.3f' % (chi2, p))
```

```python
table = pd.crosstab(cyl,gear)
from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(table.values)
print('Chi-square Statistics %0.3f p_value %0.3f' % (chi2, p))
```
