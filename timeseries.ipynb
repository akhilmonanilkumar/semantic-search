```python
import numpy as np
from numpy.random import randn
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from pylab import rcParams

import seaborn as sns
```

```python
%matplotlib inline
rcParams['figure.figsize']=5,6
sns.set_style('whitegrid')
```

```python
address = 'Superstore.csv'

df = pd.read_csv(address, index_col='Order Date', encoding='cp1252', parse_dates=True)
df.head()

```

```python
df['Order Quantity'].plot()
```

```python
df2 = df.sample(n=10, random_state=15)

plt.xlabel('Order Date')
plt.ylabel('Order Quantity')
plt.title('Superstore Sales')

df2['Order Quantity'].plot()
```

```python
address = 'mtcars.csv'

cars = pd.read_csv(address)

cars.columns = ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']

cars.index = cars.car_names

mpg = cars['mpg']

mpg.plot(kind='hist')
```

```python
plt.hist(mpg)
plt.plot()
```

```python
sns.displot(mpg)
```

```python
cars.plot(kind='scatter',x='hp', y='mpg', c=['darkgrey'], s=150)
```

```python
sns.regplot(x='hp', y='mpg', data=cars, scatter=True)
```

```python
sns.pairplot(cars)
```

```python
cars_subset = cars[['mpg','disp','hp','wt']]
sns.pairplot(cars_subset)
plt.show()
```

```python
cars.boxplot(column='mpg', by='am')
cars.boxplot(column='wt', by='am')
```

```python
sns.boxplot(x='am', y='mpg',data=cars, hue='am', palette='hls')
```
